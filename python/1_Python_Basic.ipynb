# 0. Python is an OOP language

### 1. Object . Attributes . Methods

Python is a multi-paradigm programming language, but `Object Oriented Programming` one of its core function. In python everything is an `Object`.

For everything in python, we have a `built-in class`, and an object is an example of that class. Every object, from any class, has attributes and methods. *What does it mean?*

`Class.` We have a class named Car.

`Object.` BMW is an object from that class Car. 

`Attributes.` This BMW has the follwoing attrubutes:
- Year built in: 2023
- Color: Blue
- Brand: BMW

`Methods.` This car has the following methods:
- start
- accelerate
- break
- stop

### 2. Example

```python
class car:

    # Attributes
    def __init__(self, year, color, brand, owner):
        self.year = year
        self.color = color
        self.brand = brand
        self.owner = owner

    # Methods
    def startFunc(self):
        print(f"Starting the {self.brand} car")

    def accelerateFunc(self):
        print(f"Accelerating the {self.color} {self.brand} car")

    def breakFunc(self):
        print(f"Breaking the {self.brand} car built in {self.year}")

    def stopFunc(self):
        print(f"Stopping the {self.brand} car owned by {self.owner}")
        
```

```python
BMW = car(2023, "Blue", "BMW", "Yasi")
BMW.startFunc()
BMW.accelerateFunc()
BMW.breakFunc()
BMW.stopFunc()
```

#### **`is`** vs. **`==`**

- `==` : **Same Value**
- `is` : **Exact Same Object in memory**

```python
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print(f'a == b: {a == b}')   # True  — same values
print(f'a is b: {a is b}')   # False — different objects in memory

print(f'a == c: {a == c}')   # True  — same values
print(f'a is c: {a is c}')   # True  — SAME object in memory
```

**Exception**: `Small Integers & Strings`

Python does something clever called `interning`, for small integers (usually -5 to 256) and short strings, Python **reuses the same object instead of creating a new one, to save memory**.

```python
a = 100
b = 100

print(f'a == b: {a == b}')   # True
print(f'a is b: {a is b}')   # True ← !! same object because Python cached it

print(f'id(a) == id(b): {id(a) == id(b)}')  # True
```

```python
a = 10000
b = 10000

print(f'a == b: {a == b}')   # True
print(f'a is b: {a is b}')   # False ← different objects now!
```

##### The Golden Rule

| | `==` | `is` |
|---|---|---|
| **Checks** | Same value? | Same object in memory? |
| **Uses `id()`?** | No | Yes |
| **Use it for** | Comparing values | Checking if two variables point to the same object |
| **Common use case** | `if a == b` | `if a is None` |

#### **`copy()`** vs. **`deepcopy()`**

- `copy()`: Shallow Copy, same values but different memory address.
- `deepcopy()`: a compeletly independent clone.

```python
import copy

a = [[1, 2], [3, 4]]
b = copy.copy(a)

print(f'id(a): {id(a)}')  # e.g. 1400001
print(f'id(b): {id(b)}')  # e.g. 1400002  ← different!
print(f'b: {b}')      # [1, 2, 3]     ← same values
```

```python
print(f'a: {a}')  
print(f'b: {b}')
b[0][0] = 99
print(f'a: {a}')  
print(f'b: {b}')  
```

```python
a = [[1, 2], [3, 4]]
b = copy.deepcopy(a)

print(f'id(a): {id(a)}')  # e.g. 1400001
print(f'id(b): {id(b)}')  # e.g. 1400002  ← different!
print(f'b: {b}')      # [1, 2, 3]     ← same values
```

```python
print(f'a: {a}')  
print(f'b: {b}')
b[0][0] = 99
print(f'a: {a}')  
print(f'b: {b}')  
```

#### **`a = b`** vs. **`a = copy.copy(b)`** vs. **`a = copy.deepcopy(b)`**

- **`a = b`** a is a variable that is pointing to the object b, so both of them has the same attributes, methods and reference number. So technically a is the same object as b.

- **`a = copy.copy(b)`** a is a new object, so with different reference number, but with the same ttributes and methods. However, this is a shallow copy means for nested objects, when we make any difference in a, b would be affected. So it is a coppy of outer layer, the inner layers get affected.

- **`a = copy.deepcopy(b)`** compeletly independent clone, different reference number, and if you make any changes in inner and outer layers in a, b will remain untouched. 


**`Summary Table`**

| | `b = a` | `copy.copy()` | `copy.deepcopy()` |
|---|---|---|---|
| **New object?** | ❌ No | ✅ Yes | ✅ Yes |
| **Copies outer layer?** | ❌ No | ✅ Yes | ✅ Yes |
| **Copies inner layers?** | ❌ No | ❌ No | ✅ Yes |
| **Fully independent?** | ❌ No | ⚠️ Partially | ✅ Completely |
| **Safe for nested data?** | ❌ No | ❌ No | ✅ Yes |

```python
a = [1, 2, 3]
b = a           # b is NOT a copy, it's the same object!

b.append(99)

print(f'a: {a}')  # [1, 2, 3, 99]  ← a was affected!
print(f'b: {b}')  # [1, 2, 3, 99]
```

```python
import copy

a = [1, 2, 3]
b = copy.copy(a)

print(f'id(a) == id(b): {id(a) == id(b)}')  # False ← different objects ✅

b.append(99)
print(f'a: {a}')  # [1, 2, 3]     ← a is safe ✅
print(f'b: {b}')  # [1, 2, 3, 99]
```

```python
import copy

a = [[1, 2], [3, 4]]   # a list INSIDE a list
b = copy.copy(a)

b[0][0] = 99

print(f'a: {a}')  # [[99, 2], [3, 4]]  ← a was affected! ❌
print(f'b: {b}')  # [[99, 2], [3, 4]]
```

```python
import copy

a = [[1, 2], [3, 4]]
b = copy.deepcopy(a)

b[0][0] = 99

print(f'a: {a}')  # [[1, 2], [3, 4]]   ← a is safe ✅
print(f'b: {b}')  # [[99, 2], [3, 4]]
```

### 4. Future Learning

#### The 4 Pillars of OOP

OOP is built on four big ideas. 

| Pillar | Meaning |
|---|---|
| **`Encapsulation`** | Bundling data and methods together, hiding internal details |
| **`Inheritance`** | A class can "inherit" features from another class (like a child inheriting from a parent) |
| **`Polymorphism`** | The same method name can behave differently in different classes |
| **`Abstraction`** | Hiding complex logic, showing only what's necessary |



#### Why Does This Matter?

OOP helps you write code that is:

1. **`Organized`** — related data and behavior live together
2. **`Reusable`** — write once, use in many places
3. **`Scalable`** — easier to build large programs
4. **`Maintainable`** — easier to fix and update

## Exercise 2: The User Profile Parser
You are given the following string exactly as shown:

user_data = "       alice_99 | 28 | Berlin | alice@email.com       "

Your task is to clean and extract the information from it:

1. Remove all the leading and trailing spaces from the string
2. Split the string by " | " to get each field
3. Extract the username, age, city, and email into separate variables
4. The username contains an underscore and numbers — check if the part before the underscore is all letters, and the part after the underscore is all digits
5. Build and print a formatted summary using an f-string in this exact style:

- Username: alice_99
- Age: 28
- City: Berlin
- Email: alice@email.com
- Username letters only (before underscore): True
- Username digits only (after underscore): True

**Hint**: You'll need .strip(), .split(), and string indexing to solve this.

```python
user_data = "   alice_99 | 28 | Berlin | alice@email.com   "

user_data = user_data.strip()
user_data = user_data.split(' | ')

user_name = user_data[0]
age = user_data[1]
city = user_data[2]
email = user_data[3]

print(f'Username: {user_name}')
print(f'Age: {age}')
print(f'City: {city}')
print(f'Email: {email}')
print(f'Username letters only (before underscore): {user_name.split('_')[0].isalpha()}')
print(f'Username digits only (after underscore): {user_name.split('_')[1].isdigit()}')

```

## Exercise 3: The Email Template Engine

You work at a company and need to send personalized emails to new users. You have a list of users, each represented as a raw string in this format:

users = [
    "max 24 product",
    "anna 31 engineering",
    "leo 19 design"
]

For each user, you must:

1. Parse the name, age, and department from the raw string using .split()
2. Capitalize the name properly (first letter uppercase, rest lowercase)
3. Check whether the age string is actually a valid number using .isdigit() — if it isn't, replace it with "unknown"
4. Build a personalized email using the .format() method (not an f-string) with the following template:

**pythontemplate** 

"""Hello {name},

Welcome to the {department} team! We are thrilled to have a {age}-year-old join us.
Your email has been set up as: {name_lower}@company.com

Best regards,
HR Team"""

5. Print each generated email, separated by a line of 40 dashes (- × 40)

Expected behavior for "max 24 product":

Hello Max,

Welcome to the product team! We are thrilled to have a 24-year-old join us.
Your email has been set up as: max@company.com

Best regards,
HR Team

**Hint** This exercise combines: .split(), .capitalize(), .lower(), .isdigit(), .format(), loops, and string replication ("-" * 40).

```python
email_template = """Hello {name},

Welcome to the {department} team! We are thrilled to have a {age}-year-old join us.
Your email has been set up as: {name_lower}@company.com

Best regards,
HR Team"""

users = [
    "max 24 product",
    "anna X engineering",
    "leo 19 design"
]

for user in users:
    user_info = user.split()
    user_name = user_info[0]
    user_age = user_info[1] if user_info[1].isdigit() == True else 'unknown'
    user_dept = user_info[2]

    email_personal = email_template.format(
        name = user_name.capitalize(),
        department = user_dept,
        age = user_age,
        name_lower = user_name.lower())
    
    print(email_personal)
    print(40 * '-')
```
